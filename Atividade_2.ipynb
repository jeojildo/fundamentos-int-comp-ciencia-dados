**Dois conjuntos de histogramas com tamanho de bin= 10 e bin=30**

```python
import pandas as pd
import matplotlib.pyplot as plt


df=pd.DataFrame(dadosiris)

#Criar variavel feature e guardar PetalLength
#Pegar valor unico na coluna Species e guarda na variavel species_list (usar para criar histograma baseando na class Iris)
feature = 'PetalLengthCm'
species_list = df['Species'].unique()

# Tamanho do bin usado
tamanho_bin = [10, 30]

# Criar 1 linha e 2 colunas subplot e 14,5 tamanho figura (L=14 2 A=5)
# Criar o titulo figura (f string para inserir nome do feature automaticamente)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(f"Histograma '{feature}' baseando Species", fontsize=14)

"""
for idx bin enumerate: para looping e enumerate (uma funcao para facilita a iteracao sobre uma sequencia
e necessario conhecer a posicao(index) e criar um objeto que produz pares. (criar 2 histograma com tamanho
de bin diferente)
for species in species_list: lista nome do species, para criar histograma cada species numa subplot com cores diferente.
"""
for idx, bins in enumerate(tamanho_bin):
    ax = axes[idx]
    for species in species_list:
        subset = df[df['Species'] == species]
        ax.hist(subset[feature], bins=bins, alpha=0.6, label=species, edgecolor='black')

    ax.set_title(f"{feature} dengan {bins} bins")
    ax.set_xlabel(f"{feature} (cm)")
    ax.set_ylabel("Frequencia")
    ax.legend(title="Species")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

**Meu ponto de vista:**


**Histograma, bins=10, Esquerda:**


PetalLength com bins=10, distribuicao mais global, para visao geral.

* Especie Iris-setosa(azul) o valor distribui entre 1.0 - 2.0 cm.

* Especie Iris-versicolor(laranja) o valor distribui entre 3.0 - 5.1 cm, um pouco sobreposição com virginica.

* Especie Iris-virginica(verde) o valor distribui entre 4.5 - 6.8 cm.


**Histograma, bins=30, Direita:**


PetalLength com bins=30, distribuicao mostra flutuacoes de frequencia com pequeno intervalo.

* Especie Iris-setosa(azul) o valor distribui entre 1.0 - 1.8 cm.

* Especie Iris-versicolor(laranja) o valor distribui entre 3.0 - 4.7 cm, um pouco sobreposição com virginica.

* Especie Iris-virginica(verde) o valor distribui entre 4.5 - 6.8 cm.

**Dados** : Dados-gerais e Produção-bibliográfica do Curriculo Lattes dos Professores PPGCC

**Variaveis** : Número de coautores por artigo, Número de publicações por professor ou Frequência de colaboração entre pares


**Exemplo** : Cada professor publicou pelo menos 1 artigo de pesquisa por ano (10 artigos --> 2014 a 2023)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


np.random.seed(42)

# numeros professores
num_professores = 1810

#Simulacao numeros publicacao cada professor (distribuicao Poisson com valor media=10)
publicacoes = np.random.poisson(lam=10, size=num_professores)


plt.figure(figsize=(12, 6))


sns.histplot(publicacoes, bins=20, kde=True, color='skyblue', edgecolor='black')

# Criar linha vertical para valor media
mean_val = np.mean(publicacoes)
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=1.5, label=f'Média: {mean_val:.2f}')

# Criar Label e Titulo
plt.title('Distribuição de Publicações por Professor (2014–2023)', fontsize=14)
plt.xlabel('Número de Publicações por Professor')
plt.ylabel('Quantidade de Professores')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
