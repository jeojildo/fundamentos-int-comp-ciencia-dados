# A - Correlação

```python
#import numpy
#import matplotlib
#import linregress: function para criar regressão linear(duas variáveis)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# relação linear
x1 = np.linspace(1, 9, 9)
y1 = 2 * x1 + 1

# regressão linear
slope1, intercept1, r_value1, p_value1, std_err1 = linregress(x1, y1)

# covariance
covariance1 = np.cov(x1, y1)[0, 1]

plt.figure(figsize=(6, 5))
plt.scatter(x1, y1, color='blue')
plt.plot(x1, slope1 * x1 + intercept1, color='black', linestyle='--')
plt.title("Gráfico de correlação\np-value = {:.4f}, covariance = {:.2f}".format(p_value1, covariance1))
plt.xlabel("Gene X")
plt.ylabel("Gene Y")
plt.show()
```

# B: Estimando um Intervalo de Confiança para a Média

```python
#confidence interval 90%

import numpy as np
from scipy import stats

# dados 20 estudantes
dados_valorestudante = np.array([78, 82, 90, 85, 88, 75, 92, 80, 79, 87,83, 84, 76, 89, 91, 77, 81, 86, 93, 74])

n = len(dados_valorestudante)                        # número de amostra
mean = np.mean(dados_valorestudante)                 # valor média
std_dev = np.std(dados_valorestudante, ddof=1)       # standard deviation

print("Valor média:", mean)
print("Standard deviation:", std_dev)

# intervalo de confiança de 90%
confidence = 0.90
alpha = 1 - confidence

# distribuição
t_score = stats.t.ppf(1 - alpha/2, df=n-1)

# margin of error
margin_of_error = t_score * (std_dev / np.sqrt(n))


lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Confidence interval 90%: ({lower_bound:.2f}, {upper_bound:.2f})")
```

# C: p-value em um Teste de Hipótese

```python
#efeitos do treinamento adicional nos resultados dos testes


import numpy as np
from scipy import stats

# Dados valor estudantes
grupodecontrole = np.array([78, 75, 80, 77, 76, 74, 79, 81, 73, 78])
grupodeteste = np.array([85, 87, 83, 88, 86, 84, 89, 90, 82, 88])

#Teste t independente para duas amostras
t_stat, p_value = stats.ttest_ind(grupodeteste, grupodecontrole)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
```
